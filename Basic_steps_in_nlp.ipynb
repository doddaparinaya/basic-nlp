```python
#one of the important library used for natural language processing is nltk
#importing the nltk package
import nltk
from nltk.corpus import stopwords
import re
```

```python
#reading corpus
#corpus is simply a paragraph

p='''Modi was born and raised in Vadnagar in northeastern Gujarat, where he completed his secondary education. He was introduced to the RSS at age eight. He has discussed having to work as a child in his father's tea stall in the Vadnagar railway station, which has not been reliably corroborated. At age 18, Modi was married to Jashodaben Chimanlal Modi, whom he abandoned soon after. He first publicly acknowledged her as his wife more than four decades later when required to do so by Indian law, but has made no contact with her since. Modi has asserted he had travelled in northern India for two years after leaving his parental home, visiting a number of religious centres, but few details of his travels have emerged. Upon his return to Gujarat in 1971, he became a full-time worker for the RSS. After the state of emergency was declared by prime minister Indira Gandhi in 1975, Modi went into hiding. The RSS assigned him to the BJP in 1985 and he held several positions within the party hierarchy until 2001, rising to the rank of general secretary.'''
```

```python
p
```

```python
#by using sent_tokenize ,i was converting entire corpus into documents
#documents is simply a sentence
sentences=nltk.sent_tokenize(p)
```

```python
sentences
```

```python
#from this step we get words from given corpus
words=nltk.word_tokenize(p)
```

```python
words
```

```python
len(words)
```

###stop words
####The words in nltk are the most common words in data.They are words that you do not want to see to describe the topic of your content


```python
l=stopwords.words('english')
```

```python
if "not" in l:
    print("True")
```

####while you are removing the stopwords you have to take care of some words like "not",which are used for some problems

```python
for i in sentences:
    print(i)
```

```python
#from this iam creating a new list of words without any numbers and pucntuations
new_words=[]
for i in words:
    j=re.sub('[^a-zA-Z]',"",i)
    new_words.append(j)
```

```python
new_words
```

```python
#from words i was are the common words,the words that do not describe the content that is nothing both stopwords and appending into the new list
word=[]
for i in new_words:
    if i not in set(stopwords.words('english')):
        i=i.lower()
        word.append(i)
```

```python
word
```

```python
#from all above steps i removed all common words and the numbers and punctuation marks
#so wea cn perfrom stemming and lemmitazition
```

```python
###Stemming is a process of reducing infected words to their word stem
###Advantages:This process is fast when compared to Lemmitazition
###Disadvantage:The stem word has no meaning,that is human's cannot understand
```

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
for i in word:
    j=ps.stem(i)
    print(i,"____",j)
```

```python
#The purpose of using stemming and lemmatization is same.
#Advantage:Each word is converted to meaning full word and which can be easily understandable by human
#Disadvantage:The processing converting the words to stem word is time taking process as we are converting into a meaning full word.
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lem=WordNetLemmatizer()
for i in word:
    j=lem.lemmatize(i)
    print(i,"_____",j)
```

###The major step in nlp is to convert the words to vectors because the machine only understand the numerical values.
###Previous i performed each step uniquely, now we can apply it in a simple way

```python
corpus=[]
```

```python
#this steo is simply basic preprocessing step
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

###Their are some techniques to convert strings to vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
x=cv.fit_transform(corpus).toarray()
```

```python
tfidf=TfidfVectorizer()
y=tfidf.fit_transform(corpus).toarray()
```
