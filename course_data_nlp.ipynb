####First I decided to take only reviews column

```python
y=df['course_rating_int']
dt=df.drop(columns=['course_rating_int','course_rating'],axis=1)
```

```python
d=dt['reviews']
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(d)):
    review=re.sub('[^a-zA-Z]'," ",d[i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
x=cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
m1=MultinomialNB()
m1.fit(x_train,y_train)
y_pred=m1.predict(x_test)
```

```python
from sklearn import metrics
```

```python
acc1=metrics.accuracy_score(y_pred,y_test)
matrix1=metrics.confusion_matrix(y_pred,y_test)
```

```python
print('accuracy:',acc1)
print('matrix:',matrix1)    
```
